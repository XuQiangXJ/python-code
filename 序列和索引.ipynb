# 序列和索引

 用于存放多个连续值,每个值都对应一个整数编号,可以是正向递增索引(0~N-2),也可以是反向索引(-N~1)

包含:
    1.列表
    2.元组
    3.集合
    4.字典

 列表和元组为有序序列
 集合和字典为无序序列

列表,元组,集合,字典,又被称为组合数据类型

## 1. 正向递增示例

```python
s='hello world'
for i in range(0,len(s)):
    print(i,s[i],end='\t\t')
```

## 2. 反向递增示例

```python
s='hello world'
for i in range(-len(s),0):
    print(i,s[i],end='\t\t')
```

# 2.切片操作的语法结构

    序列[start: end: step]

```python
s='helloworld'
#切片操作
s1=s[0:5:2]    #索引从0开始,到5结束,步长为2
print(s1)
print(s[:5:1])     #省略start,即从0开始
print(s[:5:])      #省略start,step,默认从0开始,步长为1
print(s[0::1])     #省略了end,默认是到最后一个终止
print(s[5::])
print(s[5:])       #第8,9行结果一致,表示省略了end和step
print(s[::2])      #默认从0开始,到最后一个索引序号,步长为2
print(s[::-1])     #默认逆序输出
print(s[-1:-11:-1])          #结果与11行一致
```
